```python
import pandas as pd

# Read the Excel file
df = pd.read_excel('data/sample.xlsx')

# Assuming the 'Block' column is named 'Block' and the 'Lab no.' column is named 'Lab no.'
block_col = 'BLOCK'
labno_col = 'Lab No'

# Combine the 'Block' and 'Lab no.' columns into a single string
df['block_labno'] = df[block_col].astype(str) + '-' + df[labno_col].astype(str)

# Get the unique values of the 'block_labno' column
unique_block_labno = df['block_labno'].unique()

# Print the unique 'block_labno' values
for block_labno in unique_block_labno:
    print(block_labno)
```

```python
import pandas as pd

# Read the CSV file
df = pd.read_csv('../data/output/output.csv')

# Create a new column that combines the desired fields in the desired format
df['Formatted Data'] = df['Date'] + ' ' +df['Class with Section'] + ' (' + df['Group (A,B,C,D)'] + ') : ' +  df['Lab Number'] + ' ' + df['Time Slot'] + ' ' + df['Internal Examiner'] + ' (' + df['External Examiner'] + ')'

# Sort the DataFrame by 'Lab Number' and 'Date'
df = df.sort_values(['Date', 'Lab Number'])

# Fetch all the values from the 'Formatted Data' column
formatted_data_values = df['Formatted Data'].values

# Print the formatted data values
for value in formatted_data_values:
    print(value)
```

```python
# Filter the dataframe based on the same time slot, lab no., and date
filtered_df = df[df['Formatted Data'].isin(formatted_data_values)]

# Get the unique class with section (group(a,b)) values
unique_class_section = filtered_df['Class with Section'].unique()

# Print the unique class with section (group(a,b)) values
for class_section in unique_class_section:
    print(class_section)
```
