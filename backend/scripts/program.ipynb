```python
import pandas as pd

# Read the CSV file
df = pd.read_csv('../data/output/output.csv')

# Parse 'Date' column as datetime
df['Date'] = pd.to_datetime(df['Date'], format='%Y-%m-%d')  # Adjust the format according to your data

# Create a new column that combines the desired fields in the desired format
df['Formatted Data'] = df['Date'].dt.strftime('%Y-%m-%d') + ' ' + df['Class with Section'] + ' (' + df['Group (A,B,C,D)'] + ') : ' + df['Lab Number'] + ' ' + df['Time Slot'] + ' ' + df['Internal Examiner'] + ' (' + df['External Examiner'] + ')'

# Sort the DataFrame by 'Date', 'Time Slot', 'Internal Examiner', and 'External Examiner'
df = df.sort_values(['Date', 'Time Slot', 'Internal Examiner', 'External Examiner'])

# Remove rows where the internal examiner is the same as the external examiner for the same time slot on the same day
df = df[~((df['Date'] == df['Date'].shift()) & (df['Time Slot'] == df['Time Slot'].shift()) & ((df['Internal Examiner'] == df['External Examiner'].shift()) | (df['External Examiner'] == df['Internal Examiner'].shift())))]

# Remove rows where the external examiner is the same for the same time slot on the same day
df = df.drop_duplicates(subset=['Date', 'Time Slot', 'External Examiner'], keep=False)

# Ensure that no unique class and section, group have more than 3 practicals per date
df = df[df.groupby(['Date', 'Class with Section', 'Group (A,B,C,D)']).cumcount() < 3]

# Ensure that no two practicals are consecutive
df['Time Slot'] = pd.to_datetime(df['Time Slot'], errors='coerce', format='%I:%M %p')  # Adjust the format according to your data
df = df.dropna(subset=['Time Slot'])  # Drop rows where parsing failed
df = df.sort_values(['Date', 'Class with Section', 'Group (A,B,C,D)', 'Time Slot'])
df = df[~(df['Time Slot'] - df['Time Slot'].shift() < pd.Timedelta(minutes=30))]

# Fetch all the values from the 'Formatted Data' column
formatted_data_values = df['Formatted Data'].values

# Print the formatted data values
for value in formatted_data_values:
    print(value)
```
